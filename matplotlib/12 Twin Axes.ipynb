when plotting curves with different units together, it is considered useful to have dual x or y axes in a figure

Matplotlib supports this with the **twinx** and **twiny** functions

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure()
a1 = fig.add_axes([0,0,1,1])
x = np.arange(1,11)
a1.plot(x,np.exp(x))
a1.set_ylabel('exp')
a2 = a1.twinx()
a2.plot(x, np.log(x),'ro-')
a2.set_ylabel('log')
fig.legend(labels = ('exp','log'),loc='upper left')
plt.show()
```

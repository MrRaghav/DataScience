```python
from matplotlib import pyplot as plt
from matplotlib import style
import numpy as np

style.use('ggplot')

x,y = np.loadtxt("C:/Users/Raghvendra/DataScience/time_series/example.csv", unpack=True, delimiter = ',')

plt.plot(x,y)

plt.title('plotting of time series')
plt.ylabel('y-axis')
plt.xlabel('x-axis')

plt.show()
```

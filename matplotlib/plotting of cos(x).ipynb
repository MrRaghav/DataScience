```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
fig, ax = plt.subplots(figsize=(10,7))

# arange for the x value with .1 as steps
x1 = np.arange(-7.5, +7.5 , .1)    #start , stop and steps

# arange for the x value with .5 as steps
x2 = np.arange(-7.5, +7.5 , .5)    #start , stop and steps

# arange for the x value with 1 as steps
x3 = np.arange(-7.5, +7.5 , 1)    #start , stop and steps

y1 = np.cos(x1)
y2 = np.cos(x2)
y3 = np.cos(x3)
# Plot the data
plt.plot(x1,y1) #smoothest with .1 step
plt.plot(x2,y2) #sharper with .5 step
plt.plot(x3,y3) #sharpest with 1 step

plt.title('example of cos function')

# Show the plot
plt.show()
```

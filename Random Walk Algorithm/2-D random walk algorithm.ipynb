*Random walk is a random process in which a walker randomly moves around some space, that is, a space with a metric used to compute distance. It is more commonly conceptualized in one dimension (Z), two dimensions (Z2) or three dimensions (Z3) in Cartesian space, where Z represents the set of integers.
Here, we have used scatter plots as well as a colorscale to denote the time sequence of the walk.

```python
import plotly.graph_objects as go
import numpy as np

l = 1000
x_steps = np.random.choice([-1, 1], size=l) + 0.5 * np.random.randn(l) # l steps
y_steps = np.random.choice([-1, 1], size=l) #+ 0.5 * np.random.randn(l) # l steps
x_position = np.cumsum(x_steps) # integrate the position by summing steps values
y_position = np.cumsum(y_steps) # integrate the position by summing steps values

fig = go.Figure(data=go.Scatter(
    x=x_position,
    y=y_position,
    mode='markers',
    name='Random Walk',
    marker=dict(
        color=np.arange(l),
        size=8,
        colorscale='PuBu',
        showscale=True
    )
))

fig.show()
```
